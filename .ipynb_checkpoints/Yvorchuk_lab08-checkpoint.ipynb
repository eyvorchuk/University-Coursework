## Part 1

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)
```

```python
a = 1
b = 4

x1 = b * np.random.rand(100)
x2 = b * np.random.rand(100)
s = np.sqrt((x1-x2)**2+a**2)
```

## Part 2

```python
left = a
right = np.sqrt(a**2+b**2)
w = 0.1
bins = np.arange(left, right, w)
```

```python
plt.hist(s, bins, label = 'histogram')
s_plot = np.arange(a + w/2, right, w/5)
pdf = (2*s_plot/b**2)*(b/(np.sqrt(s_plot**2-a**2))-1)*(100*w)
plt.plot(s_plot, pdf, label = 'pdf')
plt.title('Histogram and pdf of rod distance')
plt.xlabel('s')
plt.ylabel('frequency')
plt.legend()
```

## Part 3

```python
def negLogLikelihood(b_val):
    log_sum = 0
    for s_val in s:
        log_sum -= np.log((2*s_val/b_val**2)*(b_val/np.sqrt(s_val**2-a**2)-1))
    return log_sum
```

## Part 4

```python
b_min = np.sqrt(np.max(s)**2-a**2)
b_array = np.arange(b_min + w/10, b_min + 1, w/10)
negLog_array = np.array([negLogLikelihood(b_val) for b_val in b_array])
plt.plot(b_array, negLog_array)
plt.title('Negative Log Likelihood Function')
plt.xlabel('b')
plt.ylabel('negative log likelihood')
```

## Part 5

```python
from scipy import optimize, misc
```

```python
# From the graph, the minimum occurs between 3.8 and 4.2
result = optimize.minimize_scalar(negLogLikelihood, bounds=(3.8,4.2), method='bounded') 
b_hat = result.x

result2 = misc.derivative(negLogLikelihood, b_hat, dx=0.1, n=2)
sig_b_num = 1/np.sqrt(result2)

print("The maximum likelihood estimate for b is {0:5.3f} +/- {1:5.3f}".format(b_hat,sig_b_num))
```

## Part 6

The log likelihood function is given by

$$L(b)=\sum_{i=1}^{100}\ln{(\frac{2s_i}{b^2}(\frac{b}{\sqrt{{s_i}^2-a^2}}-1))}=\sum_{i=1}^{100}\ln{(2s_i)}-\ln{(b^2)}+\ln{(\frac{b}{\sqrt{{s_i}^2-a^2}}-1)}$$

$$\therefore\frac{dL}{db}=\sum_{i=1}^{100}\frac{-2}{b}+\frac{1}{\frac{b}{\sqrt{{s_i}^2-a^2}}-1}(\frac{1}{\sqrt{{s_i}^2-a^2}})=\sum_{i=1}^{100}\frac{-2}{b}+\frac{1}{b-\sqrt{{s_i}^2-a^2}}$$

$$\therefore\frac{d^2L}{db^2}=\sum_{i=1}^{100}\frac{2}{b^2}-\frac{1}{(b-\sqrt{{s_i}^2-a^2})^2}$$

$$\therefore\hat{\sigma}_{\hat{b}}=\sqrt{-(\frac{d^2L}{db^2})^{-1}\vert_{b=\hat{b}}}=\sqrt{\frac{-1}{\sum_{i=1}^{100}\frac{2}{{\hat{b}}^2}-\frac{1}{(\hat{b}-\sqrt{{s_i}^2-a^2})^2}}}$$

```python
secdir_sum = 0
for s_val in s:
    secdir_sum += 2/b_hat**2-1/(b_hat-np.sqrt(s_val**2-a**2))**2
sig_b_anal = np.sqrt(-1/secdir_sum)
print("The standard deviation of the estimator is {0:5.3f}".format(sig_b_anal))
print("The percent difference is {0:.2f}%.".format(abs((sig_b_anal-sig_b_num)/sig_b_anal*100)))
```

Since the percent difference is small, the values in Q5 and Q6 are close.

## Part 7

```python
b_hat_array = np.zeros(1000)
for i in range(1000):
    x1 = b * np.random.rand(100)
    x2 = b * np.random.rand(100)
    s = np.sqrt((x1-x2)**2+a**2)
    
    # Extend bounds due to more randomness
    b_min = np.sqrt(np.max(s)**2-a**2)
    result = optimize.minimize_scalar(negLogLikelihood, bounds=(b_min,6), method='bounded') 
    b_hat_array[i] = result.x
```

```python
bins = np.arange(np.min(b_hat_array), np.max(b_hat_array), 0.05)
plt.hist(b_hat_array, bins)
plt.title('Distribution of values for b-hat')
plt.xlabel('b-hat')
plt.ylabel('frequency')
```

```python
sig_dist = np.std(b_hat_array)
print("The standard deviation of the distribution is {0:5.3f}".format(sig_dist))
perc_diff_num = abs((sig_b_num-sig_dist)/sig_b_num*100)
perc_diff_anal = abs((sig_b_anal-sig_dist)/sig_b_anal*100)
print("The percent difference between this value and the one from Q5 is {0:.2f}%".format(perc_diff_num))
print("The percent difference between this value and the one from Q6 is {0:.2f}%".format(perc_diff_anal))
```

Since the standard deviation of the distribution is very close to the value in Q6, the maximum likelihood estimator is unbiased.
