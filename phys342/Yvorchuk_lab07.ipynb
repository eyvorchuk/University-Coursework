The normalization constant is obtained by integrating $g(t|H_0)$ from $-\infty$ to $\infty$. Since it's only non-zero from 0 to 1, then $\int_{0}^{1} ae^{-4t} dt = \frac{-a}{4}e^{-4t}\vert_{0}^{1} = \frac{a}{4}(1-e^{-4}) = 1$. So $a = \frac{4}{1-e^{-4}}$.

Find $t_{cut}$ such that $\int_{t_{cut}}^1\frac{4e^{-4t}}{1-e^{-4}} = 0.05$ for 95% confidence. $\int_{t_{cut}}^1\frac{4e^{-4t}}{1-e^{-4}}dt = \frac{4}{1-e^{-4}}\int_{t_{cut}}^1e^{-4t}dt = \frac{-1}{1-e^{-4}}e^{-4t}\vert_{t_{cut}}^1 = \frac{1}{1-e^{-4}}(e^{-4t_{cut}}-e^{-4}) = \frac{1}{20}$. So $e^{-4t_{cut}}-e^{-4} = \frac{1-e^{-4}}{20} => e^{-4t_{cut}} = \frac{1-e^{-4}}{20}+e^{-4} = \frac{19e^{-4}+1}{20} => t_{cut} = \frac{-1}{4}\ln{(\frac{19e^{-4}+1}{20})}$.

## (a) Transformation Method

Integrate $g(t|H_0)$ from 0 to t and equate this to r (the uniform random number). $\frac{4}{1-e^{-4}}\int_{0}^{t}e^{-4t}dt = \frac{-1}{1-e^{-4}}e^{-4t}\vert_0^t = \frac{1}{1-e^{-4}}(1-e^{-4t}) = r$. So $1-e^{-4t} = r(1-e^{-4}) => e^{-4t} = 1-r(1-e^{-4}) => t = \frac{-1}{4}\ln{(1-r(1-e^{-4}))}$.

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def nextValueForH0():
    r = random.random()
    return -0.25*np.log(1-r*(1-np.exp(-4)))
```

```python
t_cut = -0.25*np.log((19*np.exp(-4)+1)/20)
n_exp = 10000
t_arr_1 = np.zeros(n_exp)
rej_count = 0
for i in range(len(t_arr_1)):
    t_arr_1[i] = nextValueForH0()
    if(t_arr_1[i] > t_cut):
        rej_count += 1
print("Number of experiments rejecting H0: {}".format(rej_count))
```

```python
t_plot = np.linspace(0,1,10000)
pdf_1 = 4*np.exp(-4*t_plot)/(1-np.exp(-4))
plt.hist(t_arr_1, bins=20, range=(0,1), label="histogram")
plt.plot(t_plot, pdf_1*n_exp/20, label="pdf")
plt.title("Histogram and PDF if $H_0$ is true")
plt.xlabel("t")
plt.ylabel("Frequency")
plt.legend()
```

## (b) Accept/Reject Method

$sin^2(\pi{t})$ is max when $\pi{t}=\frac{\pi}{2}$, where $sin^2(\frac{\pi}{2}) = 1$. So $g_{big} = 2$.

```python
def nextValueForH1():
    tmin = 0
    tmax = 1
    while True:
        r1 = random.random()
        r2 = random.random()
        t = tmin + (tmax - tmin)*r1
        if(2*np.sin(np.pi*t)**2 > 2*r2):
            return t
```

```python
t_arr_2 = np.zeros(n_exp)
acc_count = 0
for i in range(n_exp):
    t_arr_2[i] = nextValueForH1()
    if(t_arr_2[i] <= t_cut):
        acc_count += 1
print("Number of experiments accepting H0:", acc_count)
```

```python
pdf_2 = 2*np.sin(np.pi*t_plot)**2
plt.hist(t_arr_2, bins=20, range=(0,1), label="histogram")
plt.plot(t_plot, pdf_2*n_exp/20, label="pdf")
plt.title("Histogram and PDF if $H_1$ is true")
plt.xlabel("t")
plt.ylabel("Frequency")
plt.legend()
```
