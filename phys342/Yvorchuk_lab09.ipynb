## Phys 342 Lab 9

This lab applies the least squares method to estimate parameters that are not linear model coefficients. As a result the goodness of fit test using the value of minimum $\chi^2$ is not exact.

Model: $E[V_i] = \alpha t_i \cos(\beta t_i)$

$V_i$ are gaussian distributed with standard deviation $\sigma_i$. The data is generated using random numbers with a defined seed, so the results are reproducable, if the entire notebook is rerun (Kernel -> restart & run all)

```python
# cell #1: generic setup
import numpy as np
from scipy import stats, optimize
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# cell #2: define basic parameters, the model, and create a reference dataset 

#set the seed for random number generation, so the results are reproducable
np.random.seed(seed=123456)

# define the parameters of the problem
n = 8 # number of data points
trange = [0.,4.] # range within t_i values are drawn
srange = [0.2,1.0] # corresponding range of sigma_i values
a_true = 1.5 # true value for the parameter a
b_true = 2.0 # true value for the parameter b

# Our model:
def model(a,b,t):
    return a*t*np.cos(b*t)

def get_data():
    dt = (trange[1]-trange[0])/n # distance between t values
    t = np.arange(trange[0]+dt,trange[1]+0.001,dt) # 0.001 added to ensure exactly n points in arange
    sigma = (t-trange[0])/(trange[1]-trange[0])*(srange[1]-srange[0]) + srange[0]
    v = model(a_true,b_true,t) + stats.norm.rvs(scale=sigma)
    return t, v, sigma

# Reference data:
t, v, sigma = get_data()
```

```python
# cell #3: calculate chi^2 for particular choice of parameter values

# input: args: [a,b]

def chi2(args):
    a = args[0]
    b = args[1]
    value = 0
    for i in range(n):
        value += (v[i]-a*t[i]*np.cos(b*t[i]))**2/sigma[i]**2 
    return value
```

```python
# cell #4: find best estimates for parameters using least squares for the reference dataset

# start with a guess and find the parameters that minimize the chi2 function
guess = [1.3,1.3]
# the optimize.minimize method calls the chi2 function several times to find
# parameters that produce the minimum value of that function 
# it starts from the parameters stored in guess
result = optimize.minimize(chi2,guess)
a_est = result.x[0]; b_est = result.x[1]

# the one standard deviation uncertainties are estimated by the second derivatives of the function
# since we are using chi2 instead of log-likelihood, factors of 2 needs to be included
sig_a = np.sqrt(result.hess_inv[0,0]*2.)
sig_b = np.sqrt(result.hess_inv[1,1]*2.)
rho = result.hess_inv[0,1]*2./sig_a/sig_b

print("Parameter estimates:")
print("  a-hat = %5.2f +/- %5.2f" %(result.x[0],sig_a))
print("  b-hat = %5.2f +/- %5.2f" %(result.x[1],sig_b))
print("Correlation coeff = %5.2f" %(rho))

print("chi^2 min = %5.2f" %(chi2([a_est,b_est])))
print("chi^2 true = %5.2f" %(chi2([a_true,b_true])))
```

```python
# cell #5: compare visually the data and the model

fig, axes = plt.subplots()
tarray = np.arange(trange[0],trange[1],0.02)
varray = model(a_true,b_true,tarray)
vfit = model(a_est,b_est,tarray)
axes.plot(tarray,varray,label="truth")
axes.plot(tarray,vfit,label="fit")
axes.errorbar(t,v,yerr=sigma,marker='.',linestyle='none',label='data')
plt.legend();
```

```python
# cell #6: Pull distributions from 1000 experiments

pull_a = np.zeros(1000)
pull_b = np.zeros(1000)
chi2_min = np.zeros(1000)
for i in range(1000):
    t, v, sigma = get_data()
    result = optimize.minimize(chi2, guess)
    a_est, b_est = result.x
    sig_a = np.sqrt(result.hess_inv[0,0]*2.)
    sig_b = np.sqrt(result.hess_inv[1,1]*2.)
    pull_a[i] = (a_est - a_true) / sig_a
    pull_b[i] = (b_est - b_true) / sig_b
    chi2_min[i] = chi2([a_est, b_est])

bins_a = np.linspace(np.min(pull_a), np.max(pull_a), 20)
bin_width_a = bins_a[1] - bins_a[0]
bins_b = np.linspace(np.min(pull_b), np.max(pull_b), 20)
bin_width_b = bins_b[1] - bins_b[0]
a_plot = np.linspace(np.min(pull_a), np.max(pull_a), 200)
b_plot = np.linspace(np.min(pull_b), np.max(pull_b), 200)

bins_chi = np.linspace(np.min(chi2_min), np.max(chi2_min), 20)
bin_width_chi = bins_chi[1] - bins_chi[0]
chi_plot = np.linspace(np.min(chi2_min), np.max(chi2_min), 200)

fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(15,8))

ax1.hist(pull_a, bins_a, label='Histogram')
ax1.plot(a_plot, stats.norm.pdf(a_plot, 0, 1)*1000*bin_width_a, label='Gaussian')
ax1.set_title("Pull distribution for alpha")
ax1.set_xlabel('pull_a')
ax1.set_ylabel('frequency')
ax1.legend()

ax2.hist(pull_b, bins_b, label='Histogram')
ax2.plot(b_plot, stats.norm.pdf(b_plot, 0, 1)*1000*bin_width_b, label='Gaussian')
ax2.set_title("Pull distribution for beta")
ax2.set_xlabel('pull_b')
ax2.legend()

ax3.hist(chi2_min, bins_chi, label='Histogram')
ax3.plot(chi_plot, stats.chi2.pdf(chi_plot, df=6)*1000*bin_width_chi, label='$\chi^2$')
ax3.set_title("$\chi^2_{min}$ distribution")
ax3.set_xlabel('$\chi^2_{min}$')
ax3.legend()

mean_a = np.mean(pull_a)
std_a = np.std(pull_a)
mean_b = np.mean(pull_b)
std_b = np.std(pull_b)
print("The mean and standard deviation of the first pull distribution are %5.3f and %5.3f respectively" %(mean_a, std_a))
print("The mean and standard deviation of the second pull distribution are %5.3f and %5.3f respectively" %(mean_b, std_b))
```

The mean and standard deviation of the pull distributions are close to 0 and 1 respectively because since the number of experiments is large, the central limit theorem states that the distributions for the estimated values will approximately follow Gaussian distributions. The expected values for each of these distributions are close to their respective true parameter values. Thus, subtracting these values by the true parameter value will shift the distribution to have a mean of almost 0, and dividing by the distribution's standard deviation will yield a new standard deviation that is almost 1. These are not exactly 0 and 1 respectively because the original distributions are not exactly Gaussian.

The $\chi_{min}^2$ distribution is close to the $\chi^2$ distribution with 6 degrees of freedom. Since a data point was measured every half second for four seconds, then 8 data points were taken. Since there were 2 parameters of interest, then the number of degrees of freedom for the compared $\chi^2$ distribution should be 8-2=6. Since we want to see how well the data fits the model, we compare how much the $\chi_{min}^2$ distribution matches the model $\chi^2$ distribution. In doing so, if the p-value under the assumption that the model fits the data is sufficiently large, then the assumption should be accepted.
